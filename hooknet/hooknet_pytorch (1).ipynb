## Patch Inference

```python
hooknet = HookNet(n_classes=4, n_filters=8).cuda()
hooknet.load_state_dict(torch.load("/home/user/HookNet-TIGER/last_model.pth"))
_ = hooknet.eval() 
```

```python
inference_iterator = create_batch_iterator(
    mode='inference',
    user_config=ITERATOR_CONFIG,
    cpus=4,
    iterator_class=TorchBatchIterator,
    buffer_dtype="uint8",
)
```

```python
with torch.no_grad():
    for _ in range(10):
        inputs, labels, _ = next(inference_iterator)
        t1 = time.time()
        outputs = hooknet(*inputs)
        prediction = outputs[0].detach().cpu().numpy()
        x_patch = inputs[0][0].detach().cpu().numpy()
        y_patch = labels[0][0].detach().cpu().numpy()
        print(time.time()-t1)
        
        x_patch = x_patch.transpose(1,2,0)
        x_patch = crop_data(x_patch, [1030,1030])
        y_patch = crop_data(y_patch, [1030,1030])
        plot_inference(x_patch, y_patch, prediction.transpose(1,2,0).argmax(-1))
```
